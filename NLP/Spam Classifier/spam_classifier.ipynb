## Data Preprocessing (Text Preprocessing)
- Lower case
- Tokenization
- Lemmitization
- Removing Special characters
- Removing stopwords

```python
df.head()
```

```python
def TextProcessing(text):
    text = text.lower()
    text = re.sub('[^A-Za-z0-9 ]+', '', text)
    text = [token.text for token in nlp.tokenizer(text)]
    y = []
    
    for words in text:
        if words not in (nlp.Defaults.stop_words):
           y.append(words) 
    
    y = [token.lemma_ for token in nlp(' '.join(y))]
        
    return " ".join(y)
```

```python
df['transformed_text'] = df['text'].apply(TextProcessing)
```

```python
df.head()
```

```python
# Forming word clouds to see the most frequent words
from wordcloud import WordCloud

def PlotWordCloud(target = None):
    wc = WordCloud(width = 500, height = 500, min_font_size = 10, background_color = 'white')
    spam_wc = wc.generate(df[df['target'] == target]['transformed_text'].str.cat(sep=" "))
    plt.imshow(spam_wc)
```

```python
# Plotting word cloud for "ham"
PlotWordCloud(target = 0)
```

```python
# Plotting wordcloud for "spam"
PlotWordCloud(target = 1)
```

## Transforming data to vectors | Word Embeddings

## BAG OF WORDS (BOW)

```python
# Using bag of words
cv = CountVectorizer()
X = cv.fit_transform(df['transformed_text']).toarray()
X.shape
```

```python
y = df['target'].values
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=2)
```

```python
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

```python
gnb.fit(X_train, y_train)
y_pred1 = gnb.predict(X_test)
print(accuracy_score(y_test, y_pred1))
print(confusion_matrix(y_test, y_pred1))
print(precision_score(y_test, y_pred1))
```

```python
mnb.fit(X_train, y_train)
y_pred1 = mnb.predict(X_test)
print(accuracy_score(y_test, y_pred1))
print(confusion_matrix(y_test, y_pred1))
print(precision_score(y_test, y_pred1))
```

```python
bnb.fit(X_train, y_train)
y_pred1 = bnb.predict(X_test)
print(accuracy_score(y_test, y_pred1))
print(confusion_matrix(y_test, y_pred1))
print(precision_score(y_test, y_pred1))
```
