```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
```

```python
#default size of all the graphs
plt.rcParams['figure.figsize'] = (15, 10) 
```

```python
np.random.seed(42)
x = np.random.rand(100,1) - 0.5
y = 3*x[:, 0]**2 + 0.05 * np.random.rand(100)
```

```python
df = pd.DataFrame()
```

```python
df['X'] = x.reshape(100)
df['y'] = y
```

```python
df
```

```python
plt.scatter(df['X'], df['y'])
plt.title('X vs Y')
```

# Step 1 - Gradient Boost

take mean of target variable as pred_1

```python
df['pred_1']  = df['y'].mean()
```

```python
df['res_1'] = df['y'] - df['pred_1']
```

```python
df
```

```python
plt.scatter(df['X'], df['y'])
plt.plot(df['X'], df['pred_1'], color='red')
```

# Step 2 - Train Decision tree regressor on the res_1


```python
tree1 = DecisionTreeRegressor(max_leaf_nodes = 8)
```

```python
tree1.fit(df['X'].values.reshape(100,1), df['res_1'].values)
```

```python
plot_tree(tree1)
plt.show()
```

```python
# generating X_test
X_test = np.linspace(-0.5, 0.5, 500)
```

```python
lr = 1
y_pred2 =  0.290404 + lr *(tree1.predict(X_test.reshape(500,1)))
```

```python
plt.subplot(121)
plt.plot(X_test, y_pred2, linewidth=2, color='red')
plt.scatter(df['X'], df['y'])
```

```python
df['pred_2'] = 0.265458 + tree1.predict(df['X'].values.reshape(100,1))
```

```python
df
```

```python
df['res_2'] = df['y'] - df['pred_2']
```

```python
df
```

# Step 3 - 2nd Decision Tree on the data

```python
tree2 = DecisionTreeRegressor(max_leaf_nodes=8)
```

```python
tree2.fit(df['X'].values.reshape(100,1), df['res_2'].values)
```

```python
plot_tree(tree2)
plt.show()
```

```python
y_pred3 = 0.265458 + sum(regressor.predict(X_test.reshape(-1, 1)) for regressor in [tree1,tree2])
```

```python
plt.subplot(121)
plt.plot(X_test, y_pred3, linewidth=2, color='red')
plt.scatter(df['X'], df['y'])
```
